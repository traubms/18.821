```python
def num_block_sim(N, d, T, iters):
    p = ErosionProcess(N, d)
    nb = np.array([[s.num_blocks() for s in p.run_sim(T)] for i in range(iters)])
    dist = np.array([np.histogram(nb[:, i], bins=range(d * N + 1), normed=True)[0] for i in range(nb.shape[1])])[:, ::-1].transpose()
    means = (dist * ((N * d - 1) - np.array(range(N * d)))[:, None]).sum(axis=0)
    return (nb, dist, means)
```

```python
num_blocks, dist, means = num_block_sim(10, 60, 1000, 100000)
```

```python
def calc_yhat(x, i, A, B, C, D, beta1, beta2):
    yhat = np.zeros(len(x))
    yhat[:i] = A + B*x[:i] + C * np.exp(beta1 * x[:i])
    yhat[i:] = D * np.exp(beta2 * (x[i:] - i))
    return yhat

def linear_exp(x, C, beta, A=0, B=0):
    return A + B*x + C * np.exp(x * beta)

def F(x, y, C, beta, A=0, B=0):
    yhat = linear_exp(x, C, beta, A, B)
    return np.sum(np.square(y - yhat))

def Fprime(y, i, A, C, D, beta1, beta2):
    x = np.arange(len(y))
    i = int(i)
    yhat = calc_yhat(x, i, A, 0., C, D, beta1, beta2)
    #diff_deriv  = to_pdf(y) - to_pdf(yhat)
    #diff_deriv[i-1] = 0
    return np.log(np.mean(np.abs(y - yhat)))
```

```python
W = 50
nstates = W * (W+1) / 2
T = np.zeros((nstates, nstates))
C = 0
for i in range(0, W):
    for j in range(i+1):
        if i == W-1:
            if j == 0:
                T[C, C] = 1.
            else:
                T[C, C] = .5
        else:
            T[C+i+2, C] = .5
            if j == 0:
                T[C+i+1, C] = .5
        if j > 0:
            T[C-1, C] = .5
        
        C += 1
```

```python
mapping = []
C = 0
for i in range(W):
    for j in range(i + 1):
        mapping += [(i, i-j)]
        C += 1
```

```python
def to_tree(x, mapping, W, na=False):
    tree = np.zeros((W, W)) 
    if na:
        tree *= np.nan
    for i in range(len(x)):
        p, q  = mapping[i]
        tree[p, q] = x[i]
    return tree
```
